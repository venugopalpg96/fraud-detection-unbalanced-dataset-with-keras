```python
# Split the data into X/y
X = df.drop(["Class"], axis=1).values
y = df['Class'].values
```

```python
X.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import KFold
from tensorflow.keras.optimizers import Adamax, Adam
```

```python
def create_classifier():   
    model = Sequential() 
    # First layer
    model.add(Dense(29, activation="relu"))
    model.add(Dropout(0.3))
    # hidden layer
    model.add(Dense(14, activation="relu"))
    model.add(Dropout(0.3))
    # Output layer --> Binary Classification
    model.add(Dense(1, activation="sigmoid")) 
    # Compile model
    model.compile(optimizer=Adam(learning_rate=0.0001), loss="binary_crossentropy")
    return model
```

```python
# Use train/test split for a base model
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
```

```python
ann_model = create_classifier()
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', patience=2)
```

```python
ann_model.fit(x=X_train, y=y_train, epochs=20, batch_size=64, validation_data=(X_val, y_val), callbacks=[early_stop])
```

```python
losses = pd.DataFrame(ann_model.history.history)
losses.plot()
```

```python
base_y_pred = ann_model.predict(X_val)
```

```python
base_y_pred = np.where(base_y_pred > 0.5, 1,0)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, r2_score
```

```python
print(classification_report(y_val, base_y_pred))
```

```python
confusion_matrix(y_val, base_y_pred)
```

```python
r2_score(y_val, base_y_pred)
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
# Now let't apply StratifiedKFold and see if that improve our model
skf = StratifiedKFold(n_splits=10, shuffle=True)
oof = np.zeros(len(X))
score_list = []
folds = 1

y_pred_list = []

for train_idx, test_idx in skf.split(X, y):
    X_train, X_val = X[train_idx], X[test_idx]
    y_train, y_val = y[train_idx], y[test_idx]
    
    scaler = MinMaxScaler()
    X_train = scaler.fit_transform(X_train)
    X_val = scaler.transform(X_val)
    
    early_stop = EarlyStopping(monitor='val_loss', mode='min', patience=2)
    
    ann_model = create_classifier()
    ann_model.fit(x=X_train,y=y_train,
                  validation_data=(X_val, y_val), epochs=20, batch_size=64, callbacks=[early_stop])
    
    y_preds = ann_model.predict(X_val)
    y_preds = np.where(y_preds > 0.5, 1,0)
    
    oof[test_idx] = y_preds.reshape(-1,)
    score = r2_score(y_val, oof[test_idx])
    score_list.append(score)
    print(f"Accuracy in folds {folds} : {score}")
    folds +=1
    
print(score_list)
```

```python
X = df.drop(["Class"], axis=1).values
y = df['Class'].values
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

over = SMOTE(sampling_strategy='not minority')
under = RandomUnderSampler()

steps = [("over", over),
         ("under", under)]
pipeline = Pipeline(steps=steps)

# Transform the dataset
X, y = pipeline.fit_resample(X, y)


# Apply StratifiedKFold 
skf = StratifiedKFold(n_splits=10, shuffle=True)
oof = np.zeros(len(X))
score_list = []
folds = 1

y_pred_list = []

for train_idx, test_idx in skf.split(X, y):
    X_train, X_val = X[train_idx], X[test_idx]
    y_train, y_val = y[train_idx], y[test_idx]
    
    scaler = MinMaxScaler()
    X_train = scaler.fit_transform(X_train)
    X_val = scaler.transform(X_val)
    
    early_stop = EarlyStopping(monitor='val_loss', mode='min', patience=2)
    
    ann_model = create_classifier()
    ann_model.fit(x=X_train,y=y_train,
                  validation_data=(X_val, y_val), epochs=100, batch_size=64, callbacks=[early_stop])
    
    y_preds = ann_model.predict(X_val)
    y_preds = np.where(y_preds > 0.5, 1,0)
    
    oof[test_idx] = y_preds.reshape(-1,)
    score = r2_score(y_val, oof[test_idx])
    score_list.append(score)
    print(f"Accuracy in folds {folds} : {score}")
    folds +=1
    
print(score_list)
```

```python
print(np.mean(score_list))
```

```python
print(classification_report_imbalanced(y, oof))
```
